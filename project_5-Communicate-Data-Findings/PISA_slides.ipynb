## Investigation Overview

In this investigation i wanted to have a look if the gender of a student has an impact at the score a student received in the PISA survey. 

## Dataset Overview

The original dataset contained 485490 students and 634 features. After cleaning i shortened the features to 10. Country, StudentId, InternationlGrade, BirthMonth, BirthYear, Gender, Age, MathScore, ReadingScore and ScienceScore. The scores have a range from 0 to 1000 Points.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('fivethirtyeight')


# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('pisa2012_clean.csv')
```

## Overall Distribution Of Gender

The gender distribution is almost even, with a slightly higher amount of female students.

```python
values = df_clean.Gender.value_counts()
plt.pie(values, labels= values.index, startangle=90, autopct='%1.2f%%')
plt.title('Gender Distribution')
plt.show();
```

## Overall Score Distributions

The scores are normally distributed with a mean around 500 Points. The lowest scores in each discipline are below 20 Points and the highest are over 900.

```python
fig, ax = plt.subplots(figsize=(20,5), ncols=3, nrows=1)
plt.xlim(0, 1000)

sns.distplot(df_clean.MathScore, ax=ax[0], kde=False).set_title('Math Score Distribution')
sns.distplot(df_clean.ReadingScore, ax=ax[1], kde=False).set_title('Reading Score Distribution')
sns.distplot(df_clean.ScienceScore, ax=ax[2], kde=False).set_title('Science Score Distribution');
```

## Students by Countries

Most of the countries have 50/50 gender distribution but there are some countries with a higher amount of female/male students.

```python
students_count = df_clean.Country.value_counts().sort_values(ascending=False)
plt.figure(figsize=[20,40])
sns.countplot(data=df_clean, y='Country', order=students_count.index, hue='Gender').set_title('Number of Students by Gender and Country');
```
